```python
#Joins/Merge
d={'id':np.array([1,2,3,4,5]),'Employee':np.array(['Mr. X','Mr. Y','Mr. Z','Mr. A','Mr. B'])}
```

```python
a1=pd.DataFrame(d)
a1
```

```python
d={'id':np.array([7,2,6,4,5]),'Person':np.array(['Mr. Q','Mr. R','Mr. S','Mr. T','Mr. U'])}
a2=pd.DataFrame(d)
a2
```

```python
pd.merge(a1,a2,how='left',on='id')
```

```python
pd.merge(a1,a2,how='right',on='id')
```

```python
pd.merge(a1,a2,how='inner',on='id')
```

```python
pd.merge(a1,a2,how='outer',on='id')
```
