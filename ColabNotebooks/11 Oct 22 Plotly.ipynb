```python
delivery=pd.read_csv("deliveries.csv")
match=pd.read_csv("matches.csv")
```

```python
match
```

```python
delivery.info()
```

```python
delivery.head(15).iloc[:,5:20]
```

```python
ipl=delivery.merge(match,left_on='match_id',right_on='id')
ipl
```

```python
ipl.info()
```

```python
top50=ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(50).index.tolist()
new_ipl=ipl[ipl['batsman'].isin(top50)]
new_ipl.shape
```

```python
#Calculating SR
#SR=([number of runs scored by a batsman]/[number of balls faced])*100
runs=new_ipl.groupby('batsman')['batsman_runs'].sum()
balls=new_ipl.groupby('batsman')['batsman_runs'].count()
sr=(runs/balls)*100
#sr=sr.sort_values(ascending=False)

sr=sr.reset_index()
sr
```

```python
out=ipl[ipl['player_dismissed'].isin(top50)]
nouts=out['player_dismissed'].value_counts()

avg=runs/nouts
avg=avg.reset_index()
avg.rename(columns={'index':'batsman',0:'avg'},inplace=True)
avg
```

```python
avg=avg.merge(sr,on='batsman')
avg
```

```python
avg.rename(columns={'batsman_runs':'sr'},inplace=True)
avg
```

```python
trace=go.Bar(x=avg['batsman'],y=avg['avg'])
layout=go.Layout({'title':'Average of the top 50 batsman'},
                 xaxis=dict(title='Bastmen'),
                 yaxis={'title':'Batting Average'})
data=[trace]
figure=go.Figure(data,layout)
pyo.plot(figure,filename='avg.html')
```

```python
#different options for barmode
#barmode=['stack', 'group', 'overlay', 'relative']
```

```python
trace=go.Bar(x=avg['batsman'],y=avg['avg'],name='BattingAverage',marker={'color':'green'})
trace1=go.Bar(x=avg['batsman'],y=avg['sr'],name='StrikeRate',marker={'color':'blue'})
layout=go.Layout({'title':'Average and SR comparision of the top 50 batsman in IPL history'},
                 xaxis=dict(title='Bastmen'),
                 yaxis={'title':'Batting Average/SR'},barmode='overlay')
data=[trace1,trace]
figure=go.Figure(data,layout)
pyo.plot(figure,filename='avgvssr.html')
```

```python
trace=go.Scatter(x=avg['batsman'],y=avg['avg'],name='BattingAverage',marker={'color':'black'},line={'width':5})
trace1=go.Bar(x=avg['batsman'],y=avg['sr'],name='StrikeRate',marker={'color':'red'})
layout=go.Layout({'title':'Average and SR comparision of the top 50 batsman in IPL history'},
                 xaxis=dict(title='Bastmen'),
                 yaxis={'title':'Batting Average/SR'},barmode='overlay')
data=[trace1,trace]
figure=go.Figure(data,layout)
pyo.plot(figure,filename='avgvssr.html')
```

```python
trace=go.Scatter(x=avg['avg'],y=avg['sr'],name='AvgVsSR',marker={'color':'black','size':10},mode='markers')
layout=go.Layout({'title':'Average and SR comparision of the top 50 batsman in IPL history'},
                 xaxis=dict(title='Batting Average'),
                 yaxis={'title':'StrikeRate'})
data=[trace]
figure=go.Figure(data,layout)
pyo.plot(figure,filename='avgssr.html')
```

```python
#Year by year performance of an individual batsman

single=ipl[ipl['batsman']=='V Kohli']
performance=single.groupby('season')['batsman_runs'].sum().reset_index()
performance
```

```python
single1=ipl[ipl['batsman']=='MS Dhoni']
performance1=single1.groupby('season')['batsman_runs'].sum().reset_index()
performance1
```

```python
trace=go.Scatter(x=performance['season'],y=performance['batsman_runs'],mode='lines+markers')
layout=go.Layout({'title':'Average and SR comparision of the top 50 batsman in IPL history'},
                 xaxis=dict(title='Season'),
                 yaxis={'title':'Runs Scored'})
data=[trace]
figure=go.Figure(data,layout)
pyo.plot(figure,filename='virat.html')
```
