```python
year        = [1972,    1982,     1992,     2002,     2012]
e_bangalore = [100.6,   158.61,   305.54,   394.96,   724.79]
e_mysore    = [ 10.5,   25.21,    58.65,    119.27,   274.87]
```

```python
#Colors
#r,b,g,m,c,y,k,w
#red, purple, pink, teal, orange, olive
#  #3E77B9  hex codes
```

```python
plt.figure(figsize=(14,7))
plt.plot(year,e_bangalore,color="r",label='Bangalore',marker="*",
        linestyle="--",linewidth=2,markerfacecolor='b',markeredgecolor='g' )
plt.plot(year,e_mysore,color="k",label='Mysore',marker=".",
        linestyle="-.")
plt.legend(loc="upper left",fontsize='large')
plt.xlabel("Years",fontsize=12,font='Verdana')
plt.ylabel("Electricity consumption",fontsize=12)
plt.title("Electricity consumption over the years",fontsize=15)
#plt.xticks([1972,1982,1992,2002,2012])
#plt.xticks(year)
plt.xticks(np.arange(1970,2015,2),rotation=30)
plt.yticks(np.arange(0,800,50))
plt.grid()
plt.show()
```
